### Greedy

Pseudocode:

define function:
    call the load_data function to read the city names and distance values

    create variables to track the best route and the shortest total distance

    for loop for each starting city:
        route = list starting with the current starting city
        distance_traveled = 0
        unvisited_cities = list of all city indexes except the starting city

        inner loop while there are still unvisited cities:
            current_city = last city added to the route
            call nearest_unvisited to find the closest unvisited city and its distance
            add that city to the route
            add the distance to the total distance
            remove that city from the unvisited list

        after visiting all cities:
            add distance from the last city back to the starting city

        if this routeâ€™s total distance is shorter than the current best:
            update best_route and shortest_distance with this new route

    convert best_route from index numbers to city names

    return the final route and the total distance

```python
def tsp_greedy(name_file, distance_file):
    
    # calling helper function 1
    cities, distances = read_files(name_file, distance_file)
    
    shortest_distance = LONG
    best_route = []
    
    for starting_city in range(len(cities)):
        route = [starting_city]
        distance_traveled = 0.0
        unvisited_cities = list(range(len(cities)))
        unvisited_cities.remove(starting_city)
        
        # inner loop
        while unvisited_cities:
            current_city = route[-1] # points to last item in the list
            
            # calling helper function 2
            closest_city, shortest_trip = closest_unvisited(current_city, unvisited_cities, distances)
            
            route.append(closest_city)
            distance_traveled += shortest_trip
            unvisited_cities.remove(closest_city)
            
        # returning to trip starting point
        distance_traveled += distances[route[-1]][starting_city]
        
        # updating if there is a shortest distance this time
        if distance_traveled < shortest_distance:
            shortest_distance = distance_traveled
            best_route = route
    
    final_route = [cities[i] for i in best_route]
    
    return final_route, shortest_distance
```

### Backtracking

Pseudocode:

Initialize global variables:
list routes = []
city_list = [0,1,2,3,4,5,6]

Define tsp_transaction(name_file, distance_file):
    Use global variables
    
    Call helper function read_cities(distance_file)

    For second_city range 1 to length(city_list):
        
        For last city range second_city + 1 to length(city_list) - 1:
            define partial_route = [0, second_city, last_city]
            append partial_route to routes

    Call helper function tsp_recursive(partial_route, city_list)
    print routes

```python
def tsp_backtracking(name_file, distance_file):
    global routes
    global city_list 
    read_cities(distance_file)

    for second_city in range(1,len(city_list)+1):
        for last_city in range(second_city+1, len(city_list)):
            partial_route = [0,second_city, last_city]
            routes.append(partial_route)

    tsp_recursive(partial_route, city_list)
    print(routes)
```
